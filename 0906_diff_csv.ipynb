```python
import pandas as pd
import glob
import numpy as np
pd.set_option('display.max_rows', 500)
```

```python
csv_path_list = np.array(sorted(glob.glob('./subs/*.csv')))

csv_path_list = csv_path_list[[1, 4, 5, 6, 9]].tolist()
csv_path_list
```

```python
csv_name_list = [path.rsplit("/")[2] for path in csv_path_list]
```

```python
csv_list = [pd.read_csv(path) for path in csv_path_list]
```

```python
csv_raw_id = csv_list[0]['row_id']
```

```python
csv_birds = np.array([csv_list[i]['birds'].values for i in range(len(csv_list))]).T
```

```python
csv_birds[0]
```

```python
birds_df = pd.DataFrame(csv_birds, columns=csv_name_list)
birds_df
```

```python
df_birds = pd.concat([csv_raw_id, birds_df],  axis=1)
```
