```python
# ROOT = Path.cwd().parent
# INPUT_ROOT = ROOT / "input"
INPUT_ROOT = Path("/home/knikaido/work/Cornell-Birdcall-Identification/data")
RAW_DATA = INPUT_ROOT / "birdsong-recognition"
TRAIN_AUDIO_DIR = RAW_DATA / "train_audio"
TRAIN_RESAMPLED_AUDIO_DIRS = [
  INPUT_ROOT / "birdsong-resampled-train-audio-{:0>2}".format(i)  for i in range(5)
]
TEST_AUDIO_DIR = RAW_DATA / "test_audio"
```

```python
# train = pd.read_csv(RAW_DATA / "train.csv")
train = pd.read_csv(TRAIN_RESAMPLED_AUDIO_DIRS[0] / "train_mod.csv")
train.head(5)
```

```python
sec_birds = []
for i in tqdm(range(len(train))):
    bird_list = []
    str_birds = train['secondary_labels'][i]
    if(len(str_birds)>=5):
        for text in str_birds[1:-1].split(","):
            text = re.sub(r"[,']", "", text)
#             print(text)
            if(text[0]==' '): text = text[1:]
            bird_list.append(text)
    else:
        bird_list = 'nocall'
    sec_birds.append(bird_list)
len(sec_birds)
```

```python
pri_label = train['primary_label'].unique()
ebird_codes = train['ebird_code'].unique()
```

```python
code_birds = []
for birds in sec_birds:
    bird_list = []
    if(birds!='nocall'):
        for text in birds:
            if(len(np.where(pri_label==text)[0])!=0):
                index = np.where(pri_label==text)[0][0]
                name = ebird_codes[index]
            else:
                index = -1
                name = 'other bird'
    #         print(name)
            bird_list.append(name)
    else:
        bird_list = ['nocall']
    code_birds.append(bird_list)
len(code_birds)
```

```python
codes = []
for birds in code_birds:
    bird_list = []
    if(birds!=['nocall']):
        for text in birds:
            if(text!='other bird'):
                name = BIRD_CODE[text]
                bird_list.append(name)
            else:
                bird_list.append(len(BIRD_CODE))
    else:
        bird_list = [-1]
    codes.append(bird_list)
len(codes)
```

```python
train['second_cords'] = codes
```
